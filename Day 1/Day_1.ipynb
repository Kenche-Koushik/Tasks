### Write function to calculate factorial of a number

```python
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1
    
    # Handle negative numbers
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    
    # Recursive case
    return n * factorial(n - 1)

# Test the function with some examples
print("Testing factorial function:")
print(f"factorial(0) = {factorial(0)}")
print(f"factorial(1) = {factorial(1)}")
print(f"factorial(5) = {factorial(5)}")
print(f"factorial(10) = {factorial(10)}\n")

# Test with user input
try:
    num = int(input("Enter a number to calculate its factorial: "))
    result = factorial(num)
    print(f"The factorial of {num} is {result}")
except ValueError as e:
    print(f"Error: {e}")
except RecursionError:
    print("Error: Recursion limit exceeded")
```

### Load a data set (csv) of students marks in a calls for mulitple subjects (create a csv file using ChatGPT) - 
   - Find out total marks of each student.
   - Find out average marks for each student.
   - Find out class average in each subject. 
   - Find out average marks for each class in each subject.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("students_marks.csv")
df.head()
```

```python
# Find out total marks of each student.
df["Total_Marks"] = df.iloc[:, 3:].sum(axis=1)
df.head()
```

```python
# Find out average marks for each student.
df["Average_Marks"] = df.iloc[:, 3:7].mean(axis=1)
df.head()
```

```python
# Find out class average of each class (subject)
class_average = df.groupby("Class")[df.columns[3:7]].mean()
class_average
```

```python
# Overall subject averages (for entire class)
subject_avg = df.iloc[:, 3:7].mean()
subject_avg
```
